Any signal can be decomposed into the sum of its even and odd parts:

Even part: \begin{equation*}
e[n] = \frac{1}{2}\(x[n]+x[-n]\)
\end{equation*}


Odd Part: \begin{equation*}
o[n] = \frac{1}{2}\(x[n]-x[-n]\)
\end{equation*}

and the decomposition is 
\begin{equation*}
x[n] = e[n] + o[n]
\end{equation*}


```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt



## generate a random signal
x = np.array([np.random.rand() for n in range(100)])
## and its reversed version
#neg_x = x.copy()
neg_x = x[::-1]

# display both of them
fig, ax = plt.subplots(1, 2, figsize=(8,4))
ax[0].stem(x)
ax[1].stem(neg_x)
plt.show()
```

```python
# even part of x
e_x = (x + neg_x)/2.0
# odd part of x
o_x = (x - neg_x)/2
# display both of them
fig, ax = plt.subplots(2, 2, figsize=(8,4))
ax[0][0].stem(x)
ax[0][1].stem(neg_x)
ax[1][0].stem(e_x)
ax[1][1].stem(o_x)
plt.show()
```

```python
reconstructed_x = e_x + o_x
fig, ax = plt.subplots(1, 2, figsize=(8,4))
ax[0].stem(x)
ax[1].stem(reconstructed_x)
plt.show()
```

Different (real) sinusoids

$\cos(0n)$

$\sin(0n)$

$\sin(\frac{\pi}{4}n + \frac{2\pi}{6})$

$\cos(\pi n)$

- vary the frequency and observe the effects
- take one wave at the same frequency, change the phase and observe the effects

```python
csx = [np.cos(2*np.pi*x) for x in np.linspace(0,1,10)]
#print(csx)
snx = [np.sin(2*np.pi*x) for x in np.linspace(0,1,10)]
fig, ax = plt.subplots(1, 2, figsize=(8,4))
ax[0].stem(csx)
ax[1].stem(snx)
plt.show()
```

```python
#now let's see what happens when we increase the number of samples
csx10 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,10)]
snx10 = [np.sin(2*np.pi*x) for x in np.linspace(0,1,10)]

csx20 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,20)]
snx20 = [np.sin(2*np.pi*x) for x in np.linspace(0,1,20)]

fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0][0].stem(csx10)
ax[0][1].stem(snx10)
ax[1][0].stem(csx20)
ax[1][1].stem(snx20)
plt.show()
```

Great -- this remains as one cycle, but the sampling frequency increases

Now I'll increase the frequency of the signal itself

```python
csx10f1 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,10)]
snx10f1 = [np.sin(2*np.pi*x) for x in np.linspace(0,1,10)]

csx10f2 = [np.cos(2*np.pi*2*x) for x in np.linspace(0,1,10)]
snx10f2 = [np.sin(2*np.pi*2*x) for x in np.linspace(0,1,10)]

fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0][0].stem(csx10f1)
ax[0][1].stem(snx10f1)
ax[1][0].stem(csx10f2)
ax[1][1].stem(snx10f2)
plt.show()
```

```python
csx10f1 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,10)]
csx20f1 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,20)]
csx30f1 = [np.cos(2*np.pi*x) for x in np.linspace(0,1,30)]

csx10f2 = [np.cos(2*np.pi*2*x) for x in np.linspace(0,1,10)]
csx20f2 = [np.cos(2*np.pi*2*x) for x in np.linspace(0,1,20)]
csx30f2 = [np.cos(2*np.pi*2*x) for x in np.linspace(0,1,30)]

fig, ax = plt.subplots(2, 3, figsize=(8,8))
ax[0][0].stem(csx10f1)
ax[0][1].stem(csx20f1)
ax[0][2].stem(csx30f1)
ax[1][0].stem(csx10f2)
ax[1][1].stem(csx20f2)
ax[1][2].stem(csx30f2)
plt.show()
```

Note that the first row is all one cycle, but the number of indices increase from 10, 20 to 30 as we go from left to right (increasing the sampling rate). On the second row, we have two cycles but the same increase in the number of samples

$x_1 = \cos(\frac{2\pi k}{N})$

$x_1$ is periodic with period $N$

```python
N = 10
csxkN = [np.cos(2*np.pi*k/N) for k in np.linspace(0,40,40)]
fig, ax = plt.subplots(2, 3, figsize=(12,8))
ax[0][0].stem(csxkN)
plt.show()
```

$x_1 = \cos(\frac{2\pi 3 n}{16})$

$x_1$ is periodic with period $N = 16$

```python
N = 16
csxkN = [np.cos(2*np.pi*3*n/N) for n in np.linspace(-15,15,31)]
fig, ax = plt.subplots(2, 2, figsize=(12,8))
ax[0][0].stem(csxkN)
plt.show()
```

```python
#aperiodic signal: it oscillates but is not harmonic
N = 16
csxkN = [np.cos(2*np.pi*3*n/N) for n in np.linspace(-15,15,31)]
csxkN_aperiodic = [np.cos(1.16*n) for n in np.linspace(-15,15,31)]
fig, ax = plt.subplots(2, 2, figsize=(12,8))
ax[0][0].stem(csxkN)
ax[0][1].stem(csxkN_aperiodic)
plt.show()
```
