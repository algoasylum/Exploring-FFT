```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import math
import copy 
plt.style.use('seaborn')
```

```python
t = np.linspace(0, 0.5, 400)
s = np.sin(10 * 2 * np.pi * t) + np.sin(20 * 2 * np.pi * t)
N = s.size

#im = np.full(400, 199)
im = 0.5 * ((np.sin(30 * 2 * np.pi * t) + np.sin(40 * 2 * np.pi * t)))
c_s = [complex(a,b) for a,b in zip(s ,im)]
```

```python
fft = np.fft.fft(c_s)
ifft = np.fft.ifft(fft)

T = t[1] - t[0]  # sampling interval 
# 1/T = frequency
f = np.linspace(0, 1 / T, N)

fftabs = fft.imag * 1 / N
figure, axes = plt.subplots(2,2)
plt.figure(dpi = 1500)
axes[0,0].plot(f, np.abs(fft), linewidth = 0.5)
axes[0,1].plot(f, fft.real, linewidth = 0.5)
axes[1,1].plot(f, fft.imag, linewidth = 0.5)
axes[1,0].plot(f, np.angle(fft), linewidth = 0.5)
#plt.ylabel("Amplitude")
#plt.xlabel("Frequency [Hz]")
'''
plt.plot(f, np.abs(fft), linewidth = 1.0)
plt.plot(f, fft.real, linewidth = 0.5)  #can't get why second peak won't scale
plt.plot(f, fft.imag, linewidth = 0.3)
'''
#plt.bar(f,  fft.real, width=1.5)# full wave w/o normalization factor
#plt.bar(f[:N // 2], np.abs(fft)[:N // 2] * 1 / N, width=1.5)# 1 / N is a normalization factor
#plt.bar(f[:15], fft[:15], width=1.5)
#plt.scatter(fft.real, fft.imag, linewidths = 0.0000000005)
plt.show()


print(np.angle)
```

```python
figure, axes = plt.subplots(6,6, figsize=(40,(40*2)/3), constrained_layout = True, sharex = True, sharey = True)
plt.figure(dpi = 1500)
t = np.linspace(0, 0.5, 400)
f = np.linspace(0, 1 / T, N)
for i in range(0,5,2):
  s1 = ((i/2)*10 + 10)
  s2 = ((i/2)*10 + 20)
  re = np.sin(s1 * 2 * np.pi * t) + np.sin(s2 * 2 * np.pi * t)
  
  for j in range(0,5,2):
    c1 = ((j/2)*10 + 50)
    c2 = ((j/2)*10 + 60)
    im = np.sin(c1 * 2 * np.pi * t) + np.sin(c2 * 2 * np.pi * t)
    signal = [complex(a,b) for a,b in zip(re ,im)]
    fft = np.fft.fft(signal)

    axes[i, j].plot(f, fft.real, linewidth = 0.5)
    axes[i,j].set_title('real = ' + str(s1) + ', ' + str(s2))
    axes[i, j+1].plot(f, fft.imag, linewidth = 0.5)
    axes[i, j+1].set_title('imaginary = ' + str(c1) + ', ' + str(c2))
    axes[i+1, j].plot(f, np.abs(fft), linewidth = 0.5)
    axes[i+1, j+1].stem(f, np.angle(fft)*20)
    print(s1, s2)
    print(c1, c2)
    
plt.show()
    
```
